```python
from google.cloud import aiplatform

PROJECT_ID = "dental-treatment-detection"
REGION = "us-central1"
BUCKET_NAME = f"{PROJECT_ID}-dental-data"

aiplatform.init(project=PROJECT_ID, location=REGION)

print(f"Project: {PROJECT_ID}")
print(f"Region: {REGION}")
```

```python
gcs_jsonl_uri = f"gs://{BUCKET_NAME}/dental-dataset/all_annotations.jsonl"

print(f"Creating dataset from: {gcs_jsonl_uri}")

dataset = aiplatform.ImageDataset.create(
    display_name="dental-treatment-detection",
    gcs_source=[gcs_jsonl_uri],
    import_schema_uri=aiplatform.schema.dataset.ioformat.image.bounding_box,
)

print(f"Dataset created: {dataset.display_name}")
print(f"Resource name: {dataset.resource_name}")
```
