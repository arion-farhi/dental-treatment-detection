```python
from google.cloud import aiplatform

PROJECT_ID = "dental-treatment-detection"
REGION = "us-central1"
DATASET_RESOURCE_NAME = "projects/533112800844/locations/us-central1/datasets/124866587764195328"

aiplatform.init(project=PROJECT_ID, location=REGION)

dataset = aiplatform.ImageDataset(DATASET_RESOURCE_NAME)
print(f"Loaded dataset: {dataset.display_name}")
```

```python
MODEL_DISPLAY_NAME = "dental-treatment-detector-v1"

training_job = aiplatform.AutoMLImageTrainingJob(
    display_name=MODEL_DISPLAY_NAME,
    prediction_type="object_detection",
    model_type="CLOUD_HIGH_ACCURACY_1",
)

print(f"Training job configured: {MODEL_DISPLAY_NAME}")
print(f"Model type: CLOUD_HIGH_ACCURACY_1")
```

```python
model = training_job.run(
    dataset=dataset,
    model_display_name=MODEL_DISPLAY_NAME,
    training_filter_split="labels.split=train",
    validation_filter_split="labels.split=validation",
    test_filter_split="labels.split=test",
    budget_milli_node_hours=20000,
    sync=False,
)

print(f"Training job submitted!")
print(f"Job resource name: {training_job.resource_name}")
print(f"\nMonitor progress at:")
print(f"https://console.cloud.google.com/vertex-ai/training/training-pipelines?project={PROJECT_ID}")
```
